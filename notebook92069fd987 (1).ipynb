```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../input/titanic/train.csv")
df.head(-5)
```

```python
df.drop("Name",axis="columns")
```

```python
target=df["Survived"]
inputs=df.drop("Survived",axis="columns")
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex=LabelEncoder()
le_embarked=LabelEncoder()
```

```python
inputs["Sex"]=le_sex.fit_transform(inputs["Sex"])
inputs["Embarked"] = le_embarked.fit_transform(df["Embarked"].astype(str))
inputs=inputs.drop(["Name","Ticket","Cabin"],axis="columns")
```

```python
inputs
inputs.Age=inputs.Age.fillna(inputs.Age.median())
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
model.fit(inputs,target)
```

```python
model.score(inputs,target)
```

```python
model.predict(inputs)
```
